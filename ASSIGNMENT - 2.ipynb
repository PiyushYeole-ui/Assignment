#### 5.Write a recursive function to calculate Fibonacci series up to n terms

```python
# Recursive function
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

# Fibonacci series function
def fibonacci_series(n_terms):
    series = []
    for i in range(n_terms):
        series.append(fibonacci(i))
    return series


n_terms = 10
print("Fibonacci series up to", n_terms, "terms:")
print(fibonacci_series(n_terms))
```

#### .22 Createa base Class Shape with method area().inherit Rectangle and Circle classes to inplement area calculation

```python
import math

# Base Class
class Shape:
    def area(self):
        pass


class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius



rect = Rectangle(10, 5)
print("Rectangle Area:", rect.area())

cir = Circle(7)
print("Circle Area:", cir.area())
```

#### 25. Demonstrate multiple inheritance with astudentAthlete class (inheriting from student and Athlete)

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def show_student(self):
        print("Student Name:", self.name)
        print("Grade:", self.grade)


class Athlete:
    def __init__(self, sport, rank):
        self.sport = sport
        self.rank = rank

    def show_athlete(self):
        print("Sport:", self.sport)
        print("Rank:", self.rank)



class StudentAthlete(Student, Athlete):
    def __init__(self, name, grade, sport, rank):
        Student.__init__(self, name, grade)
        Athlete.__init__(self, sport, rank)

    def show_details(self):
        print("---- Student Athlete Details ----")
        self.show_student()
        self.show_athlete()



sa = StudentAthlete("Rahul", "A", "Cricket", 1)
sa.show_details()
```

#### 27. Demonstrate opertor overloadind by implementing addition of two vector object

```python
class Complex:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, other):
        return Complex(self.real + other.real, self.imag + other.imag)

    def __str__(self):
        return f"{self.real} + {self.imag}j"

c1 = Complex(11, 12)
c2 = Complex(19,20)

result = c1 + c2
print("Result:", result)
```

#### 31.Create a class Account with private balance attribute and provide method to deposit and withdraw.

```python
class Account:
    def __init__(self, initial_balance = 0):
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Amount Deposit Successfully")
        else:
            print("Invalid deposit Amount.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print("Amount withdraw Successfully.")
        else:
            print("Insufficient balance or invalid amount.")

    def get_balance(self):
        return self.__balance

acc = Account(1000)
acc.deposit(500)
print("Current Balance:", acc.get_balance())

acc.withdraw(300)
print("Current Balance:", acc.get_balance())

acc.withdraw(2000)
print("Current Balance:", acc.get_balance())
```
